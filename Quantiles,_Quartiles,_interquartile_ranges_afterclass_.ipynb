```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Bestsellers with categories (1).csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
tenth_percentile = df['Price'].quantile(0.10)
print(f"Value that divides the lowest 10% of 'Price': {tenth_percentile}")
```

```python
age_q1=np.quantile(df['Age'],0.25)
age_q2=np.quantile(df['Age'],0.50)
age_q3=np.quantile(df['Age'],0.75)
print(f"Q1: {age_q1}\nQ2: {age_q2}\nQ3: {age_q3}")
```

```python
Q1 = df['Price'].quantile(0.25)
Q2 = df['Price'].quantile(0.50)
Q3 = df['Price'].quantile(0.75)

IQR = Q3 - Q1

print(f"1st Quartile (Q1): {Q1}")
print(f"2nd Quartile (Median, Q2): {Q2}")
print(f"3rd Quartile (Q3): {Q3}")
print(f"Interquartile Range (IQR): {IQR}")
```

```python
Q1_rating = df['User Rating'].quantile(0.25)
Q2_rating = df['User Rating'].quantile(0.50)
Q3_rating = df['User Rating'].quantile(0.75)

IQR_rating = Q3_rating - Q1_rating

print(f"1st Quartile (Q1) of User rating: {Q1_rating}")
print(f"2nd Quartile (Median, Q2) of User rating: {Q2_rating}")
print(f"3rd Quartile (Q3) of User rating: {Q3_rating}")
print(f"Interquartile Range (IQR) of User rating: {IQR_rating}")
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(y=df['Price'])
plt.title('Boxplot of Price')

plt.subplot(1, 2, 2)
sns.boxplot(y=df['User Rating'])
plt.title('Boxplot of User Rating')

plt.show()
```
