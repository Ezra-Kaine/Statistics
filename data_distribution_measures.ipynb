```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statistics as stats
```

```python
df=pd.read_csv('Bestsellers with categories.csv')
df.head()
```

```python
df.info()
```

```python
var_User_Rating=np.var(df['User Rating'])
print(f"The Variance of the User Rating is: {var_User_Rating}")
```

```python
std_User_Rating=np.std(df['User Rating'])
print(f"Standard Deviation of The User Rating is: {std_User_Rating}")
```

```python
var_Price=np.var(df['Price'])
print(f"The Variance of the Price is: {var_Price}")
```

```python
std_Price=np.std(df['Price'])
print(f"Standard Deviation of The Price is: {std_Price}")
```

```python
bins = [x * 0.5 for x in range(0, 11)]  # 0.0 â†’ 5.0

plt.figure(figsize=(8,5))
plt.hist("rating", bins=bins, edgecolor="black")
plt.xlabel("User Rating")
plt.ylabel("Frequency")
plt.title("Distribution of User Rating")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist("price", bins=bins, edgecolor="black")
plt.title("Distribution of Price")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```
