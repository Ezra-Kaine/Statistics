```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statistics as stats
```

```python
df=pd.read_csv('Bestsellers with categories.csv')
df.head()
```

```python
mean_user_rating=np.mean(df['User Rating'])
print(f"Mean of the user rating is: {mean_user_rating}")
```

```python
mean_Price=np.mean(df['Price'])
print(f"Mean of the Price is: {mean_Price}")
```

```python
mean_Reviews=np.mean(df['Reviews'])
print(f"Mean of the Reviews is: {mean_Reviews}")
```

```python
median_User_Rating=np.median(df['User Rating'])
print(f"median of the User Rating is: {median_User_Rating}")
```

```python
median_Price=np.median(df['Price'])
print(f"median of the Price is: {median_Price}")
```

```python
median_Reviews=np.median(df['Reviews'])
print(f"median of the Reviews is: {median_Reviews}")
```
