```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statistics as stats
```

```python
df=pd.read_csv('Titanic Dataset.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
age_q1=np.quantile(df['Age'],0.25)
age_q2=np.quantile(df['Age'],0.50)
age_q3=np.quantile(df['Age'],0.75)
print(f"Q1: {age_q1}\nQ2: {age_q2}\nQ3: {age_q3}")
```

```python
iqr= age_q3-age_q1
print(f"Interquantile range: {iqr}")
```

```python
plt.figure(figsize=(10,6))
bins=np.arange(1,250,20)
plt.hist(df["Fare"],bins=bins, edgecolor="blue")
plt.title("fare histogram")
plt.xlabel("Fare")
plt.ylabel("No. of passangers")
plt.grid(True,linestyle="--", alpha=0.5)
plt.show()
```
